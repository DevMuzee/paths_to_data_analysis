```python
colmeans=df.sum(numeric_only=True)/df.shape[0]
colmeans
```

```python
centered=df-colmeans #centering the numeric data
centered.describe()
```

```python
#scalling the data is by getting the standard deviation(std) of the data
scallingdata=df.std(numeric_only=True, axis=0)
cent_and_scal=centered/scallingdata #centering nd scaling a dataset
cent_and_scal.describe()
```

All the above code is done manually but there is an alternative which involve using sklearn

```python
from sklearn import preprocessing
```

```python
le=preprocessing.LabelEncoder()
df=df.apply(le.fit_transform)

scaled_data=preprocessing.scale(df)
scaled_cars=pd.DataFrame(
    scaled_data,
    index=df.index,
    columns=df.columns
)
scaled_cars.describe()
```

2. Skewed data using randomized number size

```python
normal_distributed=np.random.normal(size=10000)
normal_distributed=pd.DataFrame(normal_distributed)
normal_distributed.plot(kind='hist', figsize=(5,5), bins=30) # this is a normal dataset that is normally distributed. 
#Notice how the normally distributed data looks roughly symmetric with a bell-shaped curve
```

Generating a skewed dataset we do the following

```python
skewed=np.random.exponential(scale=2, size=10000)
skewed=pd.DataFrame(skewed)
skewed.plot(kind='hist', figsize=(5,5), bins=50)#this is called a positive or right skewed cause its moving towards the right
```

```python
#to reduce skeweness, we do the sqrt
sqrt_transformed=skewed.apply(np.sqrt)
sqrt_transformed.plot(kind='hist', figsize=(5,5), bins=50)
```

Filling missing value using imputer from sklearn

```python
from sklearn.impute import SimpleImputer

df['mpg']=np.where(df['mpg']>18, None, df['mpg'])
df['mpg']
```

```python
imp=SimpleImputer(missing_values=np.NaN,
                  strategy='mean')

imputer_range=imp.fit_transform(df)

imputer_range=pd.DataFrame(imputer_range,
                           index=df.index,
                           columns=df.columns
)

imputer_range.head(10)
```

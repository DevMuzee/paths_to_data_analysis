HISTOGRAM

```python
#histogram of the diamond carat weight
df.hist(column='carat',
        figsize=(5,5),
        color='blue'
)
```

BOXPLOT

```python
df.boxplot(column='carat')
```

```python
df.head(5)
```

```python
#One of the most useful features of a boxplot is the ability to make side-by-side boxplots.
df.boxplot(column='price',
           by='clarity', #column to split upon
           figsize=(8,8),
           color='blue'
)
```

DENSITY PLOT

```python
#more explanatory than histogram
df['carat'].plot(kind='density',
                 figsize=(5,5),
                 xlim=(0,5) #limit x axis values
)
```

```python
df.head(5)
```

```python
carat_table=pd.crosstab(index=df['clarity'], columns='count')
carat_table
```

```python
carat_table.plot(kind='bar',
                figsize=(4,4))
```

```python
clar_color=pd.crosstab(index=df['clarity'], 
                       columns=df['color'])
clar_color
```

```python
clar_color.plot(kind='bar',
                figsize=(8,8),
                stacked=True)#stacked bar plot
```

```python
clar_color.plot(kind='bar',
                figsize=(5,5),
                stacked=False)#side-by-side bar plot
```

```python
df.plot(kind='scatter',
        x='carat',
        y='price',
        figsize=(7,7),
        ylim=(0,20000)
)
```

LINE PLOT

```python
years= [y for y in range(1950, 2016)]
readings=[(y+np.random.uniform(0,20)-1900) for y in years]
time_df=pd.DataFrame({'year':years,
                      'reading':readings}
)
time_df.plot(x='year',
             y='reading',
             figsize=(9,9)
)
```

SAVING PLOTS

```python
my_plot=time_df.plot(x='year',
             y='reading',
             figsize=(9,9)
)
my_fig=my_plot.get_figure()

my_fig.savefig('LINE_PLOT_EXMP.png')

#the three ways of saving a plot are: 
#create the plot
#get the fig using .get_figure
#save the file .savefig(the name)
```
